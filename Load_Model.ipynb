```python
from pathlib import Path
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# ==========================================
# Configuration
# ==========================================

MODEL_NAME = "Qwen/Qwen2.5-Coder-7B-Instruct"

# Save inside the current notebook folder
SAVE_DIR = Path.cwd() / "qwen2.5-coder-7b"
SAVE_DIR.mkdir(parents=True, exist_ok=True)

print("=" * 70)
print(f"Downloading Model : {MODEL_NAME}")
print(f"Saving to         : {SAVE_DIR}")
print("=" * 70)

# ==========================================
# Download Tokenizer
# ==========================================

print("\nDownloading tokenizer...")

tokenizer = AutoTokenizer.from_pretrained(
    MODEL_NAME,
    trust_remote_code=True
)

print("✓ Tokenizer downloaded.")

# ==========================================
# Download Model
# ==========================================

print("\nDownloading model (This may take several minutes)...")

model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    trust_remote_code=True
)

print("✓ Model downloaded.")

# ==========================================
# Save Model
# ==========================================

print("\nSaving tokenizer...")

tokenizer.save_pretrained(SAVE_DIR)

print("Saving model...")

model.save_pretrained(SAVE_DIR)

print("\n" + "=" * 70)
print("✓ Download Complete!")
print(f"Model saved successfully at:\n{SAVE_DIR}")
print("=" * 70)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# ====================================================
# Model Path
# ====================================================

MODEL_PATH = r"C:\Users\srima\Desktop\AI\Agent\qwen2.5-coder-7b"

print("Loading tokenizer...")

tokenizer = AutoTokenizer.from_pretrained(
    MODEL_PATH,
    local_files_only=True,
    trust_remote_code=True
)

print("Loading model...")

device = "cuda" if torch.cuda.is_available() else "cpu"

model = AutoModelForCausalLM.from_pretrained(
    MODEL_PATH,
    torch_dtype=torch.float16 if device == "cuda" else torch.float32,
    local_files_only=True,
    trust_remote_code=True
).to(device)

print("\n✅ Model Loaded Successfully!")
print(f"Running on: {device}")

# ====================================================
# Chat Loop
# ====================================================

while True:

    prompt = input("\nYou: ")

    if prompt.lower() in ["exit", "quit"]:
        break

    messages = [
        {
            "role": "user",
            "content": prompt
        }
    ]

    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    inputs = tokenizer(
        text,
        return_tensors="pt"
    ).to(device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=512,
        temperature=0.7,
        do_sample=True,
        top_p=0.95
    )

    response = tokenizer.decode(
        outputs[0][inputs.input_ids.shape[1]:],
        skip_special_tokens=True
    )

    print("\nQwen:\n")
    print(response)
```
